```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import tensorflow as tf
```

```python
dataset = pd.read_excel(r'C:\Users\sanla\Desktop\Prediction of CO Conversion\Dataset_Intern.xlsx')
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
x_train = ss.fit_transform(x_train)
x_test = ss.fit_transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
ann = Sequential()
```

```python
ann.add(Dense(units = 27, activation = 'relu')) #hidden layer 1
```

```python
ann.add(Dense(units = 27, activation = 'tanh')) #hiddern layer 2
```

```python
ann.add(Dense(units = 1)) #output layer
```

```python
ann.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

```python
ann.fit(x_train, y_train, batch_size = 32, epochs = 100, verbose = 1)
```

```python
y_pred = ann.predict(x_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import r2_score, mean_squared_error, make_scorer
score = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
print(score)
print(mse)
```

```python
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.regularizers import l1, l2
model = Sequential()
model.add(Dense(units=27, activation='relu', kernel_regularizer=l2(0.001)))
model.add(Dense(units=27, activation='tanh', kernel_regularizer=l2(0.001)))
model.add(Dense(units=1))
optimizer = Adam(learning_rate=0.005)
model.compile(optimizer=optimizer, loss='mean_squared_error')
model.fit(x_train, y_train, epochs=200, batch_size=32, verbose=1)
y_pred_op = model.predict(x_test)
score_op = r2_score(y_test,y_pred_op)
print(score_op)
```

```python
mse_op = mean_squared_error(y_test, y_pred_op)
print(mse_op)
```
